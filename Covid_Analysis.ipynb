```python

df['date'] = pd.to_datetime(df.date)
df['date']
```

```python
df['year'] = pd.DatetimeIndex(df.date).year
df['month'] = pd.DatetimeIndex(df.date).month
df['day'] = pd.DatetimeIndex(df.date).day
df['weekday'] = pd.DatetimeIndex(df.date).weekday

df
```

```python
# Query the rows for May
covid_df_may = df[df.month == 5]

# Extract the subset of columns to be aggregated
covid_df_may_metrics = covid_df_may[['new_cases', 'new_deaths', 'new_tests']]

# Get the column-wise sum
covid_may_totals = covid_df_may_metrics.sum()

covid_may_totals
```

```python
covid_month_df = df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
covid_month_df
```

```python
covid_month_mean_df = df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].mean()
covid_month_mean_df 
```

```python
df['total_cases'] = df.new_cases.cumsum()
df['total_deaths'] = df.new_deaths.cumsum()
df['total_tests'] = df.new_tests.cumsum() + 393

df
```
